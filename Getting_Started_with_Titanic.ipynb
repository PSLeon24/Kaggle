```python
train_df = pd.read_csv('train.csv')
train_df.head()
```

```python
train_df.shape
```

```python
test_df = pd.read_csv("test.csv")
test_df.head()
```

```python
test_df.shape
```

```python
# The sample submission file in gender_submission.csv assumes that all female passengers survived (and all male passengers died).
women = train_df.loc[train_df.Sex == 'female']["Survived"]
rate = np.mean(women)
print(rate)
```

```python
male = train_df.loc[train_df.Sex == 'male']["Survived"]
rate2 = np.mean(male)
print(rate2)
```

From this you can see that almost 75% of the women on board survived, whereas only 19% of the men lived to tell about it. Since gender seems to be such a strong indicator of survival, the submission file in gender_submission.csv is not a bad first guess!

```python
from sklearn.ensemble import RandomForestClassifier

y = train_df["Survived"]

features = ["Pclass", "Sex", "SibSp", "Parch"] # columns
X = pd.get_dummies(train_df[features])
X_test = pd.get_dummies(test_df[features])

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X, y)
predictions = model.predict(X_test)

output = pd.DataFrame({'PassengerId': test_df.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
